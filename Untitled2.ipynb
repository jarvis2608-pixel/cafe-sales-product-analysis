```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

```python
df["Item"] = (
df["Item"] 
    .astype (str)
    .str.strip()
    .str.upper()
    .replace (["ERROR","UNKNOWN","","NaN"],np.nan)
    .fillna ("NotSpecified ")
)
```

```python
df ["Total Spent"] = pd.to_numeric(
  df["Total Spent"] 
      .astype (str)
      .str.strip(),
     errors = "coerce"
)
```

```python

df["Quantity"] = pd.to_numeric(
    df["Quantity"].astype(str).str.strip(),
    errors="coerce"
)

df["Price Per Unit"] = pd.to_numeric(
    df["Price Per Unit"].astype(str).str.strip(),
    errors="coerce"
)
```

```python
df["Payment Method"] =(
    df["Payment Method"]
    .astype(str)
    .str.strip()
    .str.upper()
    .replace(["ERRORS","UNKNOWN","","NaN"],np.nan)
    .fillna("Not Sepcified")
)
           
```

```python

df.Location=(
    df["Location"]
    .astype(str)
    .str.strip()
    .str.upper()
    .replace(["ERRORS","","NaN","UNKNOWN"],np.nan)
    .fillna("Not Specidfied")
)
```

```python
df["Transaction Date"]= pd.to_datetime( 
df["Transaction Date"].astype(str).str.strip(),
errors="coerce"
)


```

```python
df.to_csv("cleaned_dataset",index=False)
```
